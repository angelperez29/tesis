#### Funci√≥n para unir las hojas de un excel

```python
def merge_excel_file(df):
    # Diccionario con keys = sheet name y value con los datos de esa hoja
    ndf = {sheet_name: df.parse(sheet_name) for sheet_name in df.sheet_names}
    # Variables para almacenado de los conjuntos de datos
    df_result = []
    for key in ndf.keys():
        df_result.append(
            pd.DataFrame(ndf.get(key))
        )
    return df_result
```

#### Funciones para limpieza de texto

```python
def cleanText(text):
    text = re.sub(r'[^\w\s]','',text)
    text = text.lstrip()
    text = text.lstrip()
    text = text.capitalize()
    #text = emoji.replace_emoji(text, replace='')
    text = ' '.join(text.split())
    return text

clean = lambda dfReady: cleanText(dfReady) 
```

```python
def delete_hashtags(text):
    hashtag_list = re.findall('#(\w+)',text)
    for hashtag in hashtag_list:
        text = text.replace('#'+hashtag, '')
    return text

del_hash = lambda df_ready: delete_hashtags(df_ready)
```

```python
def delete_mentions(text):
    mentions_list = re.findall("@(\w+)",text)
    for mention in mentions_list:
        text = text.replace('@'+mention, '')
    return text
del_mentions = lambda df_ready: delete_mentions(df_ready)
```

```python
def delete_retwetts(text):
    text.replace(' cc', '')
    return text.replace(' RT :', '')
del_rt = lambda df_ready: delete_retwetts(df_ready)
```

```python
def delete_excess_space(text):
    text = text.strip()
    return re.sub(' +', ' ', text)
del_spaces = lambda df_ready: delete_excess_space(df_ready)
```

```python
df = pd.DataFrame(df['Tweet Text'].apply(del_hash))
df = pd.DataFrame(df['Tweet Text'].apply(del_mentions))
df = pd.DataFrame(df['Tweet Text'].apply(del_spaces))
for text in df['Tweet Text']:
    print(text + "\n")
```

```python
df = pd.DataFrame(df['Tweet Text'].apply(del_rt))
for text in df['Tweet Text']:
    print(text + "\n")
```

```python
df = pd.DataFrame(df['Tweet Text'].apply(clean))
for text in df['Tweet Text']:
    print(text + "\n")
```
