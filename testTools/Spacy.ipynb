# Prueba de herramientas

```python
import spacy as sp, pandas as pd
from nltk import SnowballStemmer
spanishstemmer=SnowballStemmer('spanish')
nlp = sp.load('es_core_news_sm')
```

```python
df = pd.read_excel('../dataset/Twitter/#_Ofensivas.xlsx')
```

```python
df = df.drop(['Link(s)','Media','Location','Verfied','Location.1','Website','Timezone','Profile Image'],axis=1)
```

## Generando tokens

```python
nlp = sp.load('es_core_news_sm')
```

```python
df['Tweet Text'] = df['Tweet Text'].astype(str)
for text in df['Tweet Text']:
    if (text!= 'nan'):
        doc = nlp(text)
        tokens = [t.orth_ for t in doc]
        print(f"""
        Text: {text.lower()}
        Tokens: {tokens}
        """)
```

## Eliminando stop words

```python
df['Tweet Text'] = df['Tweet Text'].astype(str)
for text in df['Tweet Text']:
    if (text!= 'nan'):
        doc = nlp(text)
        tokens = [t.orth_ for t in doc]
        lexical_tokens = [t.orth_ for t in doc if not t.is_punct | t.is_stop]
        print(f"""
        Text: 
        {text.lower()}
        
        Tokens: 
        {tokens}
        
        Tokens sin stop words:
        {lexical_tokens}
        """)
```

## NormalizaciÃ³n de nuestro texto

```python
def normalize(text):
    doc = nlp(text)
    words = [t.orth_ for t in doc if not t.is_punct | t.is_stop]
    lexical_tokens = [t.lower() for t in words if len(t) > 3 and     
    t.isalpha()]
    return lexical_tokens
```

```python
df['Tweet Text'] = df['Tweet Text'].astype(str)
for text in df['Tweet Text']:
    if (text!= 'nan'):
        doc = nlp(text)
        tokens = [t.orth_ for t in doc]
        lexical_tokens = [t.orth_ for t in doc if not t.is_punct | t.is_stop]
        print(f"""
        Text: 
        {text.lower()}
        
        Tokens: 
        {tokens}
        
        Tokens sin stop words:
        {lexical_tokens}
        
        Token normalizados:
        {normalize(text)}
        """)
```

```python
df['Tweet Text'] = df['Tweet Text'].astype(str)
for text in df['Tweet Text']:
    if (text!= 'nan'):
        doc = nlp(text)
        tokens = [t.orth_ for t in doc]
        lexical_tokens = [t.orth_ for t in doc if not t.is_punct | t.is_stop]
        lemmas = [tok.lemma_.lower() for tok in doc]
        print(f"""
        Text: 
        {text.lower()}
        
        Tokens: 
        {tokens}
        
        Tokens sin stop words:
        {lexical_tokens}
        
        Token normalizados:
        {normalize(text)}
        
        Token lematizados:
        {lemmas}
        """)
```

## Stemming 

```python
df['Tweet Text'] = df['Tweet Text'].astype(str)
for text in df['Tweet Text']:
    if (text!= 'nan'):
        doc = nlp(text)
        tokens = [t.orth_ for t in doc]
        lexical_tokens = [t.orth_ for t in doc if not t.is_punct | t.is_stop]
        lemmas = [tok.lemma_.lower() for tok in doc]
        tokensN = normalize(text)
        stems = [spanishstemmer.stem(token) for token in tokensN]
        print(f"""
        {'#'*100}
        Text: 
        {text.lower()}
        
        Tokens: 
        {tokens}
        
        Tokens sin stop words:
        {lexical_tokens}
        
        Token normalizados:
        {normalize(text)}
        
        Token lematizados:
        {lemmas}
        
        Stemming: 
        {stems}
        {'#'*100}
        """)                 
```
