```python
import pandas as pd
```

```python
df1 = pd.ExcelFile('../../../data/dataset_tesis/twitter/df1.xlsx')
```

```python
def mergeData(df):
    # Creamos un diccionario con cada hoja del archivo
    ndf = {sheet_name: df.parse(sheet_name) for sheet_name in df.sheet_names}
    # Variable para almacenamiento de dataframes
    array = []
    for key in ndf.keys():
        array.append(pd.DataFrame(ndf[key]))
    return array
```

```python
df1t = pd.concat(mergeData(df1))
df1t.columns
```

```python
df = df1t.drop(['Link(s)','Media','Location','Verfied','Location.1','Website','Timezone','Profile Image','Bio'],axis=1)
```

```python
df = df.dropna()
df = df.drop_duplicates(subset=['Tweet Text'])
```
